```python
class MeraLR:

    def __init__(self):    # we use constructor here 
        self.m = None
        self.b = None

    def fit(self,X_train,y_train):       

        num = 0    # we create the varibale name 'num' and 'den'                                                         
        den = 0
        
        for i in range(X_train.shape[0]):      # yeh code 160 times run krega (see below that its ('X_train.shape[0]) value is 160.)               

            num = num + ((X_train[i] - X_train.mean())*(y_train[i] - y_train.mean()))        # this is for my numerator of value 'm'
            den = den + ((X_train[i] - X_train.mean())*(X_train[i] - X_train.mean()))        # this is for my denominator of value 'm'  

            self.m = num/den
            self.b = y_train.mean() - (self.m * X_train.mean())

            print(self.m)
            print(self.b)
          
    def predict(self,X_test):

        print(X_test)       
        return  (self.m * X_test) + self.b            # as the formula is :-  y = m*x + b
                                   
```
